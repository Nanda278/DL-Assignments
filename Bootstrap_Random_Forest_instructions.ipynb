```python
x = pd.DataFrame(x)
y = pd.DataFrame(y)
sample_x =[]
sample_y=[]
x_data =[]
```

### Step 1 Creating samples: Randomly create 30 samples from the whole boston data points.

```python
j = 1
for i in range(30):
    val = random.randint(3,12)
    temp1= x.sample(n=val,axis='columns',replace=False,random_state=j)
    x_data.append(np.array(temp1))
    s = np.array(temp1.sample(n=303,random_state=j))
    temp2 = y.sample(n=303,random_state=j)
    sample_x.append(s)
    sample_y.append(temp2)
    j = j+1
print(j)
```

```python
sample_lst_x =[]
sample_lst_y =[]
```

```python
for i in range(n):
    temp1_x = pd.DataFrame(sample_x[i])
    temp1_y = pd.DataFrame(sample_y[i])
    temp2_x = temp1_x.sample(n=203,random_state=9)  #subset of each sample with 203 data pts with a particular random state
    temp2_y = temp1_y.sample(n=203,random_state=9)
    temp3_x = temp1_x
    temp3_y = temp1_y
    temp3_x = np.array(temp3_x.append(temp2_x))   #appended the subset to the same sample to create 506 data pts
    temp3_y = np.array(temp3_y.append(temp2_y))
    sample_lst_x.append(temp3_x)
    sample_lst_y.append(temp3_y)
    
```

### Step 2 Building High Variance Models on each of the sample and finding train MSE value:

```python
samples_y_pred =[]
```

```python
j=1
for i in range(30):
    tree = DecisionTreeRegressor()
    tree_reg =tree.fit(sample_lst_x[i],sample_lst_y[i])
    x_data = x.sample(n=len(sample_lst_x[i][0]),axis='columns',replace=False,random_state=j)
    y_pred = tree_reg.predict(x_data)
    samples_y_pred.append(y_pred)
    j+=1
    
```
